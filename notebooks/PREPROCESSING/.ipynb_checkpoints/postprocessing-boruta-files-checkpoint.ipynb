```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
input_folder = "/home/isadoracosenza/Documentos/ebserh-clinical-outcome/notebooks/preprocessed_files/Borutas"
output_folder = "/home/isadoracosenza/Documentos/ebserh-clinical-outcome/notebooks/preprocessed_files/Borutas_tratados"
os.makedirs(output_folder, exist_ok=True)

# Coluna específica a remover se existir
col_to_drop = 'FC__anamnese_medica_BRAQUICARDIA'

# pega todos os arquivos X_train
train_files = sorted(glob.glob(os.path.join(input_folder, "X_train*.csv")))

for train_file in train_files:

    test_file = train_file.replace("X_train", "X_test")
    
    if not os.path.exists(test_file):
        print(f"Sem o par para {train_file}")
        continue

    print(f"Processando: {train_file} | {test_file}")

    # lê os arquivos
    X_train = pd.read_csv(train_file, index_col=0)
    X_test = pd.read_csv(test_file, index_col=0)

    # drop da coluna teimosa se existir
    if col_to_drop in X_train.columns:
        X_train = X_train.drop(columns=[col_to_drop])
    if col_to_drop in X_test.columns:
        X_test = X_test.drop(columns=[col_to_drop])

    # drop colunas totalmente NaN restantes (robusto)
    X_train = X_train.loc[:, X_train.apply(lambda col: np.isfinite(col).any())]
    X_test = X_test.loc[:, X_test.apply(lambda col: np.isfinite(col).any())]

    # alinhar colunas
    common_cols = X_train.columns.intersection(X_test.columns)
    X_train = X_train[common_cols]
    X_test = X_test[common_cols]

    # salvar arquivos tratados
    base_train = os.path.basename(train_file)
    base_test = os.path.basename(test_file)
    X_train.to_csv(os.path.join(output_folder, base_train))
    X_test.to_csv(os.path.join(output_folder, base_test))

print("Arquivos tratados e salvos em:", output_folder)
```
