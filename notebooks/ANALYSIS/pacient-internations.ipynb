```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../data/dataframes-unidos-coorte2022e23.xlsx')
```

```python
list(df.columns[29:370])
```

```python
prontuario = df['prontuario']
dfpront = pd.DataFrame(prontuario)
dfpront
```

```python
def internation_counter(df, coluna_prontuario):
    # Contar quantas vezes cada prontuário aparece
    df['numero_internacoes'] = df.groupby(coluna_prontuario)[coluna_prontuario].transform('size')
    
    # Colocar 0 em todas as linhas duplicadas, exceto na primeira ocorrência
    df.loc[df.duplicated(subset=[coluna_prontuario]), 'numero_internacoes'] = ''
    
    return df
```

```python
dfprontuario = dfpront.sort_values(by='prontuario', ascending=True)

nipp = internation_counter(dfprontuario, 'prontuario')
nipp.rename(columns={'prontuario': 'prontuario_ordem'}, inplace=True)
```

```python
df_final = pd.concat([df, nipp], axis=1)
df_finall = df_final.drop('prontuario_ordem', axis=1)
coluna = df_finall.iloc[:, 305]

# Obter o nome da coluna na posição 305
numero_internacoes = df_finall.columns[305]

# Remover a coluna da posição 305
df_finall = df_finall.drop(numero_internacoes, axis=1)

# Inserir a coluna na posição 0 (início)
df_finall.insert(2, numero_internacoes, coluna)

df_finall
```

```python
df_internacoes = df_finall[['prontuario', 'numero_internacoes']]
```

```python
internacoes_por_paciente = df_internacoes.groupby(prontuario)['numero_internacoes'].count().value_counts().sort_index()
internacoes_por_paciente.plot(kind='bar', color='blue', width=0.97)
plt.xlabel('Número de internações')
plt.ylabel('Número de pacientes')
plt.yscale('log')
plt.title('Número de internações por paciente')
```
